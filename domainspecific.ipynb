```python

import os
import tarfile
import urllib.request

# Step 1: Download the dataset
url = 'http://data.vision.ee.ethz.ch/cvl/food-101.tar.gz'
data_dir = 'food-101'

if not os.path.exists(data_dir):
    print("Downloading Food-101 dataset...")
    urllib.request.urlretrieve(url, 'food-101.tar.gz')

    print("Extracting...")
    with tarfile.open('food-101.tar.gz') as tar:
        tar.extractall()
    print("Extraction complete!")

!pip install streamlit tensorflow pillow

# ========================================================
# PCOS-FRIENDLY FOOD CLASSIFIER (WITH RESNET-18)
# ========================================================

import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split

# ==========================================
# STEP 1: DEFINE VALID FOOD-101 CLASSES
# ==========================================

pcos_friendly_classes = [
    'caesar_salad', 'greek_salad', 'caprese_salad', 'miso_soup',
    'edamame', 'hummus', 'grilled_salmon', 'sushi',     # EDIT: replaced oatmeal & tomato_soup
    'chicken_curry', 'shrimp_and_grits', 'sashimi', 'seaweed_salad',
    'guacamole', 'tuna_tartare'
]

not_pcos_friendly_classes = [
    'pizza', 'donuts', 'chocolate_cake', 'cup_cakes', 'french_fries',
    'hamburger', 'hot_dog', 'ice_cream', 'apple_pie', 'cheesecake',
    'bread_pudding', 'nachos', 'pancakes', 'waffles'
]

base_images = os.path.join('food-101', 'images')
train_dir = 'data/train'
val_dir = 'data/val'

# ==========================================
# STEP 2: ORGANIZE DATA INTO TRAIN/VAL
# ==========================================

if os.path.exists('data'):
    shutil.rmtree('data')

for split in ['train', 'val']:
    for cls in ['pcos_friendly', 'not_pcos_friendly']:
        os.makedirs(os.path.join('data', split, cls), exist_ok=True)

def copy_images(classes, label_name, limit_per_class=250):
    for cls in classes:
        src_dir = os.path.join(base_images, cls)
        if not os.path.exists(src_dir):
            print(f"âš ï¸ Missing folder: {cls}")
            continue

        all_images = [os.path.join(src_dir, img) for img in os.listdir(src_dir)]
        subset = all_images[:limit_per_class]

        train_imgs, val_imgs = train_test_split(subset, test_size=0.2, random_state=42)

        for img in train_imgs:
            shutil.copy(img, os.path.join(train_dir, label_name, os.path.basename(img)))
        for img in val_imgs:
            shutil.copy(img, os.path.join(val_dir, label_name, os.path.basename(img)))

print("ğŸ“ Organizing images...")
copy_images(pcos_friendly_classes, "pcos_friendly")
copy_images(not_pcos_friendly_classes, "not_pcos_friendly")
print("âœ… Image organization complete!")


# ==========================================
# STEP 3: DATA LOADERS
# ==========================================

from tensorflow.keras.preprocessing.image import ImageDataGenerator

img_size = (224, 224)
batch_size = 32

train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.15,
    height_shift_range=0.15,
    zoom_range=0.25,
    shear_range=0.15,
    horizontal_flip=True,
    brightness_range=[0.7, 1.3],
    fill_mode='nearest'
)

val_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_data = val_gen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary',
    shuffle=False                 # EDIT: ensure CM aligns with labels
)

print("Class indices:", train_data.class_indices)


# ========================================================
# STEP 4: MODEL 2 â€” RESNET-18 (CUSTOM IMPLEMENTATION)
# ========================================================

import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, ReLU, Add, MaxPooling2D
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Input, Dropout  # EDIT: added Dropout
from tensorflow.keras.models import Model

def residual_block(x, filters, stride=1):
    shortcut = x

    # Main path
    x = Conv2D(filters, (3,3), strides=stride, padding='same')(x)
    x = BatchNormalization()(x)
    x = ReLU()(x)

    x = Conv2D(filters, (3,3), strides=1, padding='same')(x)
    x = BatchNormalization()(x)

    # Adjust shortcut if shape mismatches
    if stride != 1 or shortcut.shape[-1] != filters:
        shortcut = Conv2D(filters, (1,1), strides=stride, padding='same')(shortcut)
        shortcut = BatchNormalization()(shortcut)

    x = Add()([x, shortcut])
    x = ReLU()(x)
    return x


def build_resnet18(input_shape=(224,224,3)):
    inputs = Input(shape=input_shape)

    x = Conv2D(64, (7,7), strides=2, padding='same')(inputs)
    x = BatchNormalization()(x)
    x = ReLU()(x)
    x = MaxPooling2D((3,3), strides=2, padding='same')(x)

    # ResNet layers
    x = residual_block(x, 64)
    x = residual_block(x, 64)

    x = residual_block(x, 128, stride=2)
    x = residual_block(x, 128)

    x = residual_block(x, 256, stride=2)
    x = residual_block(x, 256)

    x = residual_block(x, 512, stride=2)
    x = residual_block(x, 512)

    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.4)(x)   # helps prevent overfitting
    outputs = Dense(1, activation='sigmoid')(x)

    model = Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model


# Build model
resnet18 = build_resnet18()
resnet18.summary()


# ========================================================
# STEP 5: TRAIN RESNET-18
# ========================================================

print("ğŸŸ£ Training ResNet-18...")
history_resnet = resnet18.fit(
    train_data,
    validation_data=val_data,
    epochs=25
)

loss, acc = resnet18.evaluate(val_data)
print(f"âœ… Final Validation Accuracy: {acc*100:.2f}%")


# ========================================================
# STEP 6: SAVE MODEL
# ========================================================

resnet18.save("pcos_resnet18_model.h5")
print("ğŸ’¾ Saved as pcos_resnet18_model.h5")


# ========================================================
# STEP 7: SINGLE IMAGE PREDICTION
# ========================================================

from tensorflow.keras.preprocessing import image

def predict_food_image(img_path, model, threshold=0.5):
    img = image.load_img(img_path, target_size=(224,224))
    x = image.img_to_array(img) / 255.0
    x = np.expand_dims(x, axis=0)

    pred = float(model.predict(x)[0][0])
    label = "ğŸ¥— PCOS-friendly" if pred >= threshold else "ğŸŸ Not PCOS-friendly"

    print(f"{label} ({pred:.2f})")
    return label, pred

# ============================================================
# ğŸŒ¿ CALOCARE â€” PCOS-FRIENDLY FOOD CLASSIFIER (Streamlit + Colab)
# ============================================================

!pip install -q streamlit pyngrok tensorflow pillow

import os
from pyngrok import ngrok

# ğŸ” Replace with your real ngrok token
NGROK_AUTH_TOKEN = "34zUvtdzaMxee0ts3sQ5VuQn3RL_75WrRvEfMntQZLjnX4MX5"
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# ------------------------------------------
# ğŸ’¾ Write the Streamlit app code
# ------------------------------------------
app_code = """
import streamlit as st
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
from PIL import Image

# ------------------------------------------
# ğŸŒ¿ CALOCARE - PAGE CONFIG
# ------------------------------------------
st.set_page_config(page_title='CaloCare | PCOS Food Classifier', layout='centered')

st.markdown(
    '<h1 style="text-align:center; color:#2E8B57;">ğŸŒ¿ CaloCare</h1>',
    unsafe_allow_html=True
)
st.markdown(
    '<h4 style="text-align:center;">Your Smart Nutrition Companion for PCOS Wellness</h4><br>',
    unsafe_allow_html=True
)

st.write('Upload a food image to check if it is **PCOS-Friendly** or not. ğŸ´')

# ------------------------------------------
# ğŸ§  LOAD MODEL
# ------------------------------------------
@st.cache_resource
def load_model():
    model = tf.keras.models.load_model('pcos_resnet18_model.h5', compile=False)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = load_model()

# ------------------------------------------
# ğŸ“¸ IMAGE UPLOAD + PREDICTION
# ------------------------------------------
uploaded_file = st.file_uploader('ğŸ“¤ Upload a food image', type=['jpg', 'jpeg', 'png'])

if uploaded_file:
    img = Image.open(uploaded_file).convert('RGB')
    img_resized = img.resize((224, 224))
    st.image(img, caption='ğŸ–¼ï¸ Uploaded Image', use_column_width=True)

    img_array = image.img_to_array(img_resized) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    with st.spinner('Analyzing your food... ğŸ½ï¸'):
        prediction = model.predict(img_array)[0][0]

    label = 'ğŸŒ¿ PCOS-Friendly' if prediction > 0.5 else 'ğŸŸ Not PCOS-Friendly'
    conf = float(prediction) if prediction <= 1 else 1.0

    st.markdown('---')
    st.markdown(
        f"<h3 style='text-align:center; color:#2E8B57;'>Prediction: {label}</h3>",
        unsafe_allow_html=True
    )
    st.progress(conf)
    st.markdown(f"<p style='text-align:center;'>Confidence: {conf:.2f}</p>", unsafe_allow_html=True)

    # Recommendations
    if prediction > 0.5:
        st.success('âœ… Great choice! This food supports a PCOS-friendly diet.')
        st.write('**Tip:** Continue incorporating such nutrient-rich, low-GI foods.')
    else:
        st.error('âš ï¸ Try to limit this food in your diet.')
        st.write('**Tip:** Consider grilled proteins, veggies, or low-sugar options instead.')

st.markdown('<br><hr><center>Â© 2025 CaloCare | Empowering Health through Smart AI</center>', unsafe_allow_html=True)
"""

# Save the app code
with open("app.py", "w") as f:
    f.write(app_code)

# ------------------------------------------
# ğŸš€ Launch the Streamlit app
# ------------------------------------------
if not os.path.exists("pcos_resnet18_model.h5"):
    raise FileNotFoundError("âŒ Please upload your model file 'pcos_friendly_food_model.h5' first!")

print("ğŸš€ Launching CaloCare app...")

ngrok.kill()
!streamlit run app.py --server.port 8501 &>/dev/null &

public_url = ngrok.connect(8501)
print(f"âœ… CaloCare is live! Visit your app here:\n\n{public_url}")

import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Reset the generator (important)
val_data.reset()

# True labels
y_true = val_data.classes

# Predictions
y_pred_probs = resnet18.predict(val_data, verbose=1)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)

# Display labels in correct order
labels = list(train_data.class_indices.keys())

# Plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap=plt.cm.Blues, values_format='d')
plt.title("Confusion Matrix - ResNet18")
plt.show()

import matplotlib.pyplot as plt

# -----------------------
# ACCURACY PLOT
# -----------------------
plt.figure(figsize=(8,5))
plt.plot(history_resnet.history['accuracy'], label='Train Accuracy')
plt.plot(history_resnet.history['val_accuracy'], label='Val Accuracy')
plt.title("Accuracy Curve")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------
# LOSS PLOT
# -----------------------
plt.figure(figsize=(8,5))
plt.plot(history_resnet.history['loss'], label='Train Loss')
plt.plot(history_resnet.history['val_loss'], label='Val Loss')
plt.title("Loss Curve")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```
