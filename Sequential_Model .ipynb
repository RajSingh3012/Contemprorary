```python
# ===== Week-5 Time-Series Script (RNN vs LSTM vs GRU) — minimal edits for sequence-length effects =====
# - Reads: /content/Weather/weather.csv  (fallback to /mnt/data/weather.csv)
# - Tests multiple sequence lengths: SEQ_LENS = [24, 48, 72]
# - Same pipeline, gradient clipping, inline plots, predictions de-normalized to °C
# - Adds MAE-vs-window and Time-vs-window summary plots

import numpy as np
import pandas as pd
import torch, torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import matplotlib.pyplot as plt
import time, re, os

# -----------------------------
# 1) Load CSV and pick target (same as before, with fallback path)
# -----------------------------
CSV_PATHS = ["/content/Weather/weather.csv", "/mnt/data/weather.csv"]
for p in CSV_PATHS:
    if os.path.exists(p):
        CSV_PATH = p
        break
else:
    raise FileNotFoundError("weather.csv not found. Tried /content/Weather/weather.csv and /mnt/data/weather.csv")

df = pd.read_csv(CSV_PATH)

# Find a temperature column robustly (fallback = first numeric col)
num_cols = [c for c in df.columns if pd.api.types.is_numeric_dtype(df[c])]
cand = [c for c in df.columns if re.search(r"temp", c, re.I)] or \
       [c for c in df.columns if re.search(r"temperature", c, re.I)]
target_col = cand[0] if cand else (num_cols[0] if num_cols else None)
assert target_col is not None, f"No numeric columns found in {CSV_PATH}."
temps_all = df[target_col].astype("float32").to_numpy()
print(f"Using temperature column: {target_col} | N={len(temps_all)} | file={CSV_PATH}")

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# -----------------------------
# 2) Models (unchanged)
# -----------------------------
HID      = 128
LAYERS   = 2
HORIZON  = 1
BATCH    = 64
EPOCHS   = 20
LR       = 1e-3
CLIP     = 1.0

class SeqPredictor(nn.Module):
    def __init__(self, kind="RNN"):
        super().__init__()
        if kind == "RNN":
            core = nn.RNN(1, HID, LAYERS, batch_first=True, nonlinearity="tanh")
        elif kind == "LSTM":
            core = nn.LSTM(1, HID, LAYERS, batch_first=True)
        else:
            core = nn.GRU(1, HID, LAYERS, batch_first=True)
        self.kind = kind
        self.core = core
        self.head = nn.Linear(HID, HORIZON)

    def forward(self, x):
        out, _ = self.core(x)           # (B, L, H)
        last = out[:, -1, :]            # (B, H)
        return self.head(last).squeeze(1)  # (B,)

def train_model(model, tr_dl, va_dl, epochs=EPOCHS):
    model = model.to(DEVICE)
    opt = optim.Adam(model.parameters(), lr=LR)
    mse = nn.MSELoss()

    tr_hist, va_hist = [], []
    t0 = time.time()
    for ep in range(1, epochs+1):
        model.train(); tr_loss=0; n=0
        for xb, yb in tr_dl:
            xb, yb = xb.to(DEVICE), yb.to(DEVICE)
            opt.zero_grad()
            pred = model(xb)
            loss = mse(pred, yb)
            loss.backward()
            nn.utils.clip_grad_norm_(model.parameters(), CLIP)
            opt.step()
            tr_loss += loss.item()*xb.size(0); n += xb.size(0)
        tr_loss /= n

        model.eval(); va_loss=0; m=0
        with torch.no_grad():
            for xb, yb in va_dl:
                xb, yb = xb.to(DEVICE), yb.to(DEVICE)
                va_loss += mse(model(xb), yb).item()*xb.size(0); m += xb.size(0)
        va_loss /= m

        tr_hist.append(tr_loss); va_hist.append(va_loss)
        print(f"[{model.kind}] ep {ep:02d}/{epochs} | train {tr_loss:.4f} | val {va_loss:.4f}")

    secs = time.time() - t0
    return {"model": model, "train": tr_hist, "val": va_hist, "secs": secs}

def mae_degC(y_true_deg, y_pred_deg):
    return float(np.mean(np.abs(y_true_deg - y_pred_deg)))

def predict_denorm(model, Xn, y_mean, y_std):
    model.eval()
    with torch.no_grad():
        y_hat_n = model(torch.from_numpy(Xn).to(DEVICE)).cpu().numpy()
    return y_hat_n * y_std + y_mean

# -----------------------------
# 3) Minimal: run the SAME pipeline for multiple seq lengths
# -----------------------------
SEQ_LENS = [24, 48, 72]   # <— minimal new line to test sequence-length effects

results = {L: {} for L in SEQ_LENS}

for seq_len in SEQ_LENS:
    print(f"\n=== seq_len = {seq_len} ===")
    # --- windows
    def make_xy(series, L=seq_len, H=1):
        X, y = [], []
        for i in range(len(series) - L - H + 1):
            X.append(series[i:i+L])
            y.append(series[i+L:i+L+H][0])
        X = np.array(X, dtype="float32")[:, :, None]  # (N, L, 1)
        y = np.array(y,  dtype="float32")
        return X, y

    X_all, y_all = make_xy(temps_all, L=seq_len, H=1)

    # --- chronological split
    n_train = int(0.8 * len(X_all))
    X_tr, X_va = X_all[:n_train], X_all[n_train:]
    y_tr, y_va = y_all[:n_train], y_all[n_train:]

    # --- normalize X and y (train stats only)
    x_mean, x_std = X_tr.mean(), X_tr.std() + 1e-8
    y_mean, y_std = y_tr.mean(), y_tr.std() + 1e-8

    X_tr_n = (X_tr - x_mean) / x_std
    X_va_n = (X_va - x_mean) / x_std

    y_tr_n = (y_tr - y_mean) / y_std
    y_va_n = (y_va - y_mean) / y_std

    tr_ds = TensorDataset(torch.from_numpy(X_tr_n), torch.from_numpy(y_tr_n))
    va_ds = TensorDataset(torch.from_numpy(X_va_n), torch.from_numpy(y_va_n))
    tr_dl = DataLoader(tr_ds, batch_size=BATCH, shuffle=True, drop_last=True)
    va_dl = DataLoader(va_ds, batch_size=BATCH, shuffle=False)

    # --- train all three
    arts = {}
    for kind in ["RNN", "LSTM", "GRU"]:
        print(f"\nTraining {kind} …")
        arts[kind] = train_model(SeqPredictor(kind), tr_dl, va_dl)

    # --- loss curves for THIS window (inline)
    plt.figure(figsize=(10,5))
    for k, c in zip(["RNN","LSTM","GRU"], ["tab:blue","tab:green","tab:purple"]):
        plt.plot(arts[k]["train"], label=f"{k} Train", color=c)
        plt.plot(arts[k]["val"],   label=f"{k} Val",   linestyle="--", color=c)
    plt.xlabel("Epoch"); plt.ylabel("MSE (normalized)"); plt.title(f"Training & Validation Loss — seq_len={seq_len}")
    plt.legend(); plt.show()

    # --- MAE (°C) + time for summary
    for k in ["RNN","LSTM","GRU"]:
        y_hat_deg = predict_denorm(arts[k]["model"], X_va_n, y_mean, y_std)
        mae = mae_degC(y_va, y_hat_deg)   # y_va is raw degrees
        results[seq_len][k] = {"mae": mae, "secs": arts[k]["secs"]}
        print(f"{k:4s} | Val MAE: {mae:.3f} °C | Time: {arts[k]['secs']:.1f}s")

    # --- (show one forecast plot only for seq_len==48 to avoid spamming)
    if seq_len == 48:
        best = min(["RNN","LSTM","GRU"], key=lambda k: results[seq_len][k]["mae"])
        best_model = arts[best]["model"]
        y_hat_va_deg = predict_denorm(best_model, X_va_n, y_mean, y_std)
        Nshow = min(300, len(y_va))
        plt.figure(figsize=(12,4))
        plt.plot(y_va[:Nshow], label="True")
        plt.plot(y_hat_va_deg[:Nshow], label=f"{best} Forecast")
        plt.xlabel("Time Step"); plt.ylabel("Temperature (°C)")
        plt.title(f"Forecast vs True (val) — seq_len={seq_len} | Best={best}")
        plt.legend(); plt.show()

# -----------------------------
# 4) Summary: MAE vs window & Time vs window (inline)
# -----------------------------
models = ["RNN","LSTM","GRU"]
windows = SEQ_LENS

# MAE vs window
plt.figure(figsize=(8,4))
for k in models:
    plt.plot(windows, [results[L][k]["mae"] for L in windows], marker="o", label=k)
plt.xlabel("Sequence length (lookback)"); plt.ylabel("Val MAE (°C)")
plt.title("MAE vs Sequence Length")
plt.legend(); plt.show()

# Time vs window
plt.figure(figsize=(8,4))
for k in models:
    plt.plot(windows, [results[L][k]["secs"] for L in windows], marker="o", label=k)
plt.xlabel("Sequence length (lookback)"); plt.ylabel("Training time (s)")
plt.title("Training Time vs Sequence Length")
plt.legend(); plt.show()

# Also print a tiny table for your report
print("\n=== Summary (Val MAE °C | Time s) ===")
for L in windows:
    row = " | ".join([f"{k}: {results[L][k]['mae']:.3f}°C, {results[L][k]['secs']:.1f}s" for k in models])
    print(f"L={L:>3} -> {row}")
```
